```python
from urllib.parse import urljoin
```

```python
coin_name = []
coin_abb = []
url_add = []
list_soup = soup.find_all('td', 'no-wrap currency-name')

for item in list_soup:
    coin_name.append(item.find(class_='currency-name-container link-secondary').get_text())
    coin_abb.append(item.find(class_='currency-symbol visible-xs').get_text())
    url_add.append(urljoin(url, item.find('a')['href']))
```

```python
data = {'Coin_Name':coin_name, 'Coin_Abb':coin_abb, 'URL':url_add}
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('1227_coinurl10.csv', sep=',', encoding='UTF-8')
```
